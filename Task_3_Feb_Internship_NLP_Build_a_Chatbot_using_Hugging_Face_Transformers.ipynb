```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration
# Importing the Model and Tokenizer
tokenizer = BlenderbotTokenizer.from_pretrained("facebook/blenderbot-400M-distill")
model = BlenderbotForConditionalGeneration.from_pretrained("facebook/blenderbot-400M-distill")
# load the BlenderBot 400M distilled model from Hugging Face’s model hub

while True:
    user_input = input("User: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Have a great day.")
        break

    guided_input = "You are a helpful assistant. " + user_input # role prompt(tells the model: “You are a helpful assistant,” so it behaves more like an AI assistant )

    inputs = tokenizer([guided_input], return_tensors="pt") # text is converted into numerical format

    reply_ids = model.generate(**inputs)
    bot_reply = tokenizer.decode(reply_ids[0], skip_special_tokens=True) # numerical output is converted back into human‑readable text
    print(f"Chatbot: {bot_reply}")
```
